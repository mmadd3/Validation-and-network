```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
import math

%pylab inline
```

```python
data=pd.read_csv("breast_cancer.csv") #here we are importing the brest cancer dateset
```

```python
y = data.pop("malignant") #this is the data we wanted to predict
```

```python
data.head()
```

in the next step i am going to drop the data that seems to be not important

```python
data.drop(["Unnamed: 0", "id number"], axis=1)
```

```python
categorical_variables = ['uniformity_of_cell_shape', 'uniformity_of_cell_size']
for variable in categorical_variables:
    dummies = pd.get_dummies(data[variable], prefix=variable)
    data = pd.concat([data, dummies], axis=1)
    data.drop([variable], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=.2, random_state=42)
```

```python
### Grid Search
n_estimators = [300,400,500]
max_features = ['auto', 'sqrt','log2']
min_samples_split = [3,5,7]


rfc = RandomForestClassifier(n_jobs=1)
#Parameters of pipelines can be set using ‘__’ separated parameter names:
estimator = GridSearchCV(rfc,
                         dict(n_estimators=n_estimators,
                              max_features=max_features,
                              min_samples_split=min_samples_split
                              ), cv=None, n_jobs=-1)
                              
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
best_rfc = estimator.best_estimator_
```

# Accuracy

```python
accuracy = accuracy_score(y_test, best_rfc.predict(X_test))
print "Accuracy: ", accuracy
```

# Precision and recall

```python
print classification_report(y_test, best_rfc.predict(X_test))
```

Precision: The model will correctly identify 96% of malignant tumour.   Conversely there is a 4% chance it will identify a non-malignant as a malignant.  (tp / tp+fp)

Recall:  Of the people the model identified as malignant, 98% of them will actually be malignant.  (tp / tp + fn)

```python
print confusion_matrix(y_test, best_rfc.predict(X_test))
```

**Precision:**  44 / (44 + 2) = .96 (fp=2)

**Recall:**   44 / (44 + 1) = .98 (fn=1)
